```python
with open('/content/average_squad_age_raw.txt', 'r') as f:
  info = f.readlines()
  for i in info:
    print(i)
```

```python
for i in range(len(info)):
  info[i] = info[i].strip()
```

```python
for i in range(len(info)):
  if info[i] == '':
    info.pop(i)
```

```python
indices = [i  for i in range(0, len(info), 3)]
```

```python
info = [info[i] for i in range(len(info)) if i not in indices]
```

```python
info
```

```python
teams = []
avg_age = []

for i in range(len(info)):
  if i % 2 == 0:
    teams.append(info[i])
  else:
    avg_age.append(float(info[i]))
```

```python
teams[0] = 'Czechia'
teams[1] = 'Türkiye'
```

```python
avg_age
```

```python
teams
```

```python
df_ = pd.DataFrame({'Team': teams,
                    'Average Age': avg_age})
```
