```python
import pandas as pd
```

```python
lines = []
with open('/content/fifa_rankings_raw.txt', 'r') as f:
  for line in f:
    lines.append(line.strip())
```

```python
def is_integer(s):
  for char in s:
    if not char.isdigit():
      return False

  return True
```

```python
lines = lines[1:]
lines
```

```python
ele_dict = {ele: is_integer(ele) for ele in lines}
```

```python
keys_l = list(ele_dict.keys())

for key_idx in range(len(keys_l)):
  key = keys_l[key_idx]
  if ele_dict[key]:
    ele_dict[keys_l[key_idx + 1]] = False
```

```python
info = [key for key, val in ele_dict.items() if (not is_integer(key)) or (is_integer(key) and ele_dict[key])]
```

```python
info
```

```python
def get_indices(lst_length, start_idx, step):

  indices = []
  idx = start_idx

  while idx < lst_length:
    indices.append(idx)
    idx += step

  return indices
```

```python
ranking_indices = get_indices(len(info), 0, 3)
rankings = [int(ele) for i, ele in enumerate(info) if i in ranking_indices]
rankings
```

```python
team_indices = get_indices(len(info), 1, 3)
teams = [ele for i, ele in enumerate(info) if i in team_indices]
teams
```

```python
total_points_l = []
prev_points_l = []
delta_points_l = []

point_indices = get_indices(len(info), 2, 3)

for idx in point_indices:
  data = info[idx].split('\t')

  total_points_l.append(float(data[0]))
  prev_points_l.append(float(data[1]))
  delta_points_l.append(float(data[2]))
```

```python
df = pd.DataFrame({'FIFA Ranking': rankings,
                   'Team': teams,
                   'Total Points': total_points_l,
                   'Previous Points': prev_points_l,
                   'Change in Points': delta_points_l})
```

```python
df.to_csv('fifa_rankings_uefa_june_2024.csv', index=False)
```
