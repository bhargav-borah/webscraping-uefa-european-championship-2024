```python
import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
import requests
```

```python
URL = 'https://en.wikipedia.org/wiki/UEFA_Euro_2024'

response = requests.get(URL)
soup = BeautifulSoup(response.content, 'html5lib')
```

```python
# soup
```

```python
tbodies = soup.find_all('tbody')
tbodies
```

```python
table = tbodies[4]
```

```python
nations = []
base_camps = []
training_grounds = []

for i in range(1, len(table.find_all('tr'))):
  data = table.find_all('tr')[i].find_all('td')
  nations.append(data[0].find('a').text)
  base_camps.append(data[1].find('a').text)
  try:
    training_grounds.append(data[2].find('a').text)
  except:
    training_grounds.append(data[2].text.strip())
```

```python
df = pd.DataFrame({'Team': nations,
                   'Base Camp': base_camps,
                   'Training Ground': training_grounds})
df
```

```python
df.to_csv('team-base_camp-training_ground.csv', index=False)
```
