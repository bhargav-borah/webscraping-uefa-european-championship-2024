```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_v1.csv')
```

```python
df['Group'] = ''
```

```python
def assign_group(row):
  if row['Team'] in ['Germany', 'Switzerland', 'Scotland', 'Hungary']:
    return 'Group A'

  elif row['Team'] in ['Spain', 'Italy', 'Albania', 'Croatia']:
    return 'Group B'

  elif row['Team'] in ['England', 'Denmark', 'Slovenia', 'Serbia']:
    return 'Group C'

  elif row['Team'] in ['Netherlands', 'France', 'Austria', 'Poland']:
    return 'Group D'

  elif row['Team'] in ['Romania', 'Slovakia', 'Belgium', 'Ukraine']:
    return 'Group E'

  else:
    return 'Group F'
```

```python
df['Group'] = df.apply(assign_group, axis=1)
```
